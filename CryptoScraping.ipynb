```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
website = "https://www.coingecko.com/en"
```

```python
response = requests.get(website)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
print(soup)
```

```python
results = soup.find('table', {'class':'sort table mb-0 text-sm text-lg-normal table-scrollable'}).find('tbody').find_all('tr')
len(results)
```

```python
#adding data to lists
name = []
price = []
chg_1h = []
chg_24h = []
chg_7d = []
vol_24h = []
mkt_cap = []
listoflists = [name,price,chg_1h,chg_24h,chg_7d,vol_24h,mkt_cap]

for result in results:
  #name
  try:
    name.append(result.find('a', {'class':'tw-hidden lg:tw-flex font-bold tw-items-center tw-justify-between'}).get_text().strip())
  except:
    name.append('n/a')  
  #price 
  try:
    price.append(result.find('td', {'class':'td-price'}).get_text().strip())
  except:
    price.append('n/a') 
  #1h change
  try:
    chg_1h.append(result.find('td', {'class':'td-change1h'}).get_text().strip())
  except:
    chg_1h.append('n/a') 
  #24h change
  try:
   chg_24h.append(result.find('td', {'class':'td-change24h'}).get_text().strip())
  except:
    chg_24h.append('n/a')
  #7d change
  try:
    chg_7d.append(result.find('td', {'class':'td-change7d'}).get_text().strip())
  except:
    chg_7d.append('n/a') 
  #24h volume
  try:
    vol_24h.append(result.find('td', {'class':'td-liquidity_score'}).get_text().strip())
  except:
    vol_24h.append('n/a')
    #market cap
  try:
    mkt_cap.append(result.find('td', {'class':'td-market_cap'}).get_text().strip())
  except:
    mkt_cap.append('n/a') 

for x in listoflists:
  print(x)
  
```

```python
crypto_df = pd.DataFrame({"Coin":name, "Price":price, "1 hour change": chg_1h, "24 hour change":chg_24h, "7 day change":chg_7d, "24 hour volume":vol_24h, "Market Cap":mkt_cap})
crypto_df.head()
```
